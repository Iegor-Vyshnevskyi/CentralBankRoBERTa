```python
bis = pd.read_pickle('/content/drive/MyDrive/CentralBankBERTa/Audience_classification/data_input/BIS_Labelled_No_Cutoff.csv')
```

```python
print(int(len(bis)))
bis['Predicted Audience'].value_counts()
```

```python
households = bis[bis['Predicted Audience'] == 0].reset_index(drop=True)
firms = bis[bis['Predicted Audience'] == 1].reset_index(drop=True)
banks = bis[bis['Predicted Audience'] == 2].reset_index(drop=True)
government = bis[bis['Predicted Audience'] == 3].reset_index(drop=True)
cb = bis[bis['Predicted Audience'] == 4].reset_index(drop=True)
```

```python
households_084 = households
index = 0 
N_length = int(len(households_084))
for i in range(0, N_length): 
  if max(households_084["Audience (Probabilities)"][index]) <= 0.84: # 2268 for 0.84, 638 forr 0.85
         households_084['Predicted Audience'][index] = "n"
  index = index + 1
households_084['Predicted Audience'].value_counts()
```

```python
firms_067 = firms
index = 0 
N_length = int(len(firms_067))
for i in range(0, N_length): 
  if max(firms_067["Audience (Probabilities)"][index]) <= 0.67: #1645 for 0.67, 595 for 0.68 
         firms_067['Predicted Audience'][index] = "n"
  index = index + 1
firms_067['Predicted Audience'].value_counts()
```

```python
banks_092 = banks
index = 0 
N_length = int(len(banks_092))
for i in range(0, N_length): 
  if max(banks_092["Audience (Probabilities)"][index]) <= 0.92: #0.9 114247, 0.91: 56141, 0.92: 7614
         banks_092['Predicted Audience'][index] = "n"
  index = index + 1
banks_092['Predicted Audience'].value_counts()
```

```python
government_071 = government
index = 0 
N_length = int(len(government_071))
for i in range(0, N_length): 
  if max(government_071["Audience (Probabilities)"][index]) <= 0.71: #0.72 485 
         government_071['Predicted Audience'][index] = "n"
  index = index + 1
government_071['Predicted Audience'].value_counts()
```

```python
cb_78 = cb
index = 0 
N_length = int(len(cb_78))
for i in range(0, N_length): 
  if max(cb_78["Audience (Probabilities)"][index]) <= 0.78: #0.9: nothing, 0.85: nothing, 0.8: 3, 0.78:1114, 79:107   
         cb_78['Predicted Audience'][index] = "n"
  index = index + 1
cb_78['Predicted Audience'].value_counts()
```

```python
h = households_084[households_084['Predicted Audience'] != 'n'].reset_index(drop=True)
f = firms_067[firms_067['Predicted Audience'] != 'n'].reset_index(drop=True)
b = banks_092[banks_092['Predicted Audience'] != 'n'].reset_index(drop=True)
g = government_071[government_071['Predicted Audience'] != 'n'].reset_index(drop=True)
cb = cb_78[cb_78['Predicted Audience'] != 'n'].reset_index(drop=True)
```

```python
# take samples here
h_bis = h.sample(1000)
f_bis = f.sample(750)
b_bis = b.sample(1500)
g_bis = g.sample(750)
cb_bis = cb.sample(804)
```

```python
bis_labelled = pd.concat([h_bis,f_bis,b_bis,g_bis,cb_bis]).reset_index(drop=True)
```

```python
bis_labelled = bis_labelled.drop(['Audience (Probabilities)'], axis=1)
bis_labelled = bis_labelled.rename(columns={'text': 'Parsed_Text', 'Predicted Audience': 'label'})
bis_labelled
```
